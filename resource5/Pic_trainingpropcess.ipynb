```python
import json
import matplotlib.pyplot as plt
from matplotlib.legend_handler import HandlerLine2D
import matplotlib
import numpy as np
x_label = "FM"
y_label = "Average Noise"
Batchsize = 128
matplotlib.rcParams['font.family'] = 'Liberation Serif'
# 替换为实际的 JSON 文件路径
loss_history_path = '1_model/Loss_Thrust_model.json'

# 读取损失历史
with open(loss_history_path, 'r') as f:
    loss_history = json.load(f)
train_loss_history = np.array(loss_history["train_loss"])/Batchsize
valid_loss_history = np.array(loss_history["valid_loss"])/Batchsize

# 绘制损失曲线
plt.figure(figsize=(12, 8))

line1, = plt.plot(train_loss_history, label='Training Loss', color='blue')
line2, = plt.plot(valid_loss_history, label='Validation Loss', color='red')

line1, = plt.plot(train_loss_history, marker='o', markersize = 2,linestyle='-', color='blue')
line2, = plt.plot(valid_loss_history, marker='v', markersize = 2, linestyle='--', color='red')

# 设置图表标题和坐标轴标签
plt.title('Training and Validation Loss (Log Scale)', fontsize=14)
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Loss (log scale)', fontsize=12)

# 设置 y 轴为对数刻度
plt.yscale('log')

# 添加网格线，可以更好地估计数值
plt.grid(True, which="both", linestyle='--', linewidth=0.5)


# 添加图例
plt.legend()

plt.savefig("training loss.pdf")
# 显示图表
plt.show()
```
